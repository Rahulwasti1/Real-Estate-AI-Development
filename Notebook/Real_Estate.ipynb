```python
df = df.drop(columns={'Views','Posted'}) # Removing columns
```

### `For 'Title' column`

```python
a = df['Title'].unique()

for i in a:
    print(i)
```

```python
df['Title'] = df['Title'].replace(r'[^a-zA-Z0-9, ]', '', regex=True)
```

```python
property_types = ['House', 'Land', 'Apartment', 'Business', 'Flat', 'Office Space', 'Villa', 'Hostel', 'Bungalow']

df['Property_Types'] = df['Title'].str.extract(f"({'|'.join(property_types)})", expand=False)

df
```
